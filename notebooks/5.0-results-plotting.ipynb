```python
import numpy as np
diamos = {
    "Task":  ["Disease Detection", "Symptom Identification", "Combined Task"],
    "Accuracy": [0.9837, 0.8853, 0.8666],
    "F1-Score": [0.9918, 0.8919, 0.8705]
}

plantvillage = {
    "Task":  ["Disease Detection", "Symptom Identification", "Combined Task"],
    "Accuracy": [0.9856, 0.9305, 0.9478],
    "F1-Score": [0.9901, 0.9302, 0.9475]
}

plantdoc = {
    "Task":  ["Disease Detection", "Symptom Identification", "Combined Task"],
    "Accuracy": [0.8392, 0.7569, 0.7754],
    "F1-Score": [0.8954, 0.7692, 0.7897]
}

megaplant = {
    "Task":  ["Disease Detection", "Symptom Identification", "Combined Task"],
    "Accuracy": [0.9807, 0.9386, 0.9512],
    "F1-Score": [0.98, 0.95, 0.95]
}


diamos_acc_mean = np.mean(diamos["Accuracy"])
diamos_f1_mean = np.mean(diamos["F1-Score"])
plantvillage_acc_mean = np.mean(plantvillage["Accuracy"])
plantvillage_f1_mean = np.mean(plantvillage["F1-Score"])
plantdoc_acc_mean = np.mean(plantdoc["Accuracy"])
plantdoc_f1_mean = np.mean(plantdoc["F1-Score"])
megaplant_acc_mean = np.mean(megaplant["Accuracy"])
megaplant_f1_mean = np.mean(megaplant["F1-Score"])


print("DiaMOS - Accuracy Mean:", diamos_acc_mean)
print("DiaMOS - F1-Score Mean:", diamos_f1_mean)
print("PlantVillage - Accuracy Mean:", plantvillage_acc_mean)
print("PlantVillage - F1-Score Mean:", plantvillage_f1_mean)
print("PlantDoc - Accuracy Mean:", plantdoc_acc_mean)
print("PlantDoc - F1-Score Mean:", plantdoc_f1_mean)
print("MegaPlant - Accuracy Mean:", megaplant_acc_mean)
print("MegaPlant - F1-Score Mean:", megaplant_f1_mean)
```

```python
fd_datasets_acc_mean = np.mean([diamos_acc_mean, plantvillage_acc_mean, plantdoc_acc_mean])
fd_datasets_f1_mean = np.mean([diamos_f1_mean, plantvillage_f1_mean, plantdoc_f1_mean])

print("Foundational Datasets - Accuracy Mean:", fd_datasets_acc_mean)
print("Foundational Datasets - F1-Score Mean:", fd_datasets_f1_mean)
```
